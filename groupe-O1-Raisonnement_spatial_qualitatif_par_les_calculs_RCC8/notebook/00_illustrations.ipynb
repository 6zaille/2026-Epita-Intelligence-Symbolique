```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def draw_circle(ax, x, y, r, label):
    circle = Circle((x, y), r, fill=False)
    ax.add_patch(circle)
    ax.text(x, y, label, ha='center', va='center')

fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()

# 1. DC (Disconnected)
ax = axes[0]
draw_circle(ax, 1, 1, 0.5, "A")
draw_circle(ax, 3, 1, 0.5, "B")
ax.set_title("DC (Disconnected)")
ax.set_xlim(0, 4)
ax.set_ylim(0, 2)

# 2. EC (Externally Connected)
ax = axes[1]
draw_circle(ax, 1, 1, 0.7, "A")
draw_circle(ax, 2.4, 1, 0.7, "B")
ax.set_title("EC (Touching)")
ax.set_xlim(0, 4)
ax.set_ylim(0, 2)

# 3. PO (Partial Overlap)
ax = axes[2]
draw_circle(ax, 1.5, 1, 0.8, "A")
draw_circle(ax, 2.2, 1, 0.8, "B")
ax.set_title("PO (Overlap)")
ax.set_xlim(0, 4)
ax.set_ylim(0, 2)

# 4. EQ (Equal)
ax = axes[3]
draw_circle(ax, 2, 1, 0.8, "A = B")
ax.set_title("EQ (Equal)")
ax.set_xlim(0, 4)
ax.set_ylim(0, 2)

# 5. TPP (A dans B, touche bord)
ax = axes[4]
draw_circle(ax, 2, 1, 1.0, "B")
draw_circle(ax, 2.6, 1, 0.5, "A")
ax.set_title("TPP (A inside B, touches border)")
ax.set_xlim(0, 4)
ax.set_ylim(0, 2)

# 6. NTPP (A dans B, pas touche bord)
ax = axes[5]
draw_circle(ax, 2, 1, 1.2, "B")
draw_circle(ax, 2, 1, 0.5, "A")
ax.set_title("NTPP (A inside B)")
ax.set_xlim(0, 4)
ax.set_ylim(0, 2)

plt.tight_layout()
plt.show()
```
